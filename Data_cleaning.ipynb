# Remove Corrupted Images where all pixels values are zeros

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/TEZDA_TASK/Dataset/age_gender.csv')

# Create a function to check if all pixel values are zero
def is_all_zero(image_pixels):
    pixel_values = np.fromstring(image_pixels, dtype=int, sep=' ')
    return np.all(pixel_values == 0)

# Check for images with all-zero pixel values and create a mask
zero_intensity_mask = df['pixels'].apply(is_all_zero)

# Filter out images with all-zero pixel values
cleaned_df = df[~zero_intensity_mask]

# Print the number of removed images
num_removed_images = sum(zero_intensity_mask)
print(f'Removed {num_removed_images} images with all-zero pixel values.')

# Now 'cleaned_df' contains the dataset without images with all-zero pixel values
```

# Find the missing age values

```python

expected_ages = set(range(1, 116))

# Get the unique age values from your dataset
unique_ages = set(df['age'].unique())

# Find the missing age values
missing_ages = expected_ages - unique_ages

# Check if there are any missing ages
if len(missing_ages) > 0:
    print("Missing age values:", missing_ages)
else:
    print("No missing age values in the dataset")
```
