```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
csv_file_path = pd.read_csv('/content/drive/MyDrive/TEZDA_TASK/Dataset/age_gender.csv')
# dataset = pd.read_csv(csv_file_path)
# Assuming df is your dataset DataFrame
# Create a DataFrame with age frequencies
age_counts = csv_file_path['age'].value_counts().sort_index().reset_index()
age_counts.columns = ['Age', 'Frequency']









```

```python

# Sample Image Display
sample_indices = [0, 1, 2]  # Indices of samples to display
plt.figure(figsize=(12, 4))
for i, idx in enumerate(sample_indices):
    plt.subplot(1, len(sample_indices), i + 1)
    pixel_string = csv_file_path.loc[idx, 'pixels']
    pixel_values = [int(val) for val in pixel_string.split()]
    image_data = np.array(pixel_values).reshape(48, 48)
    plt.imshow(image_data, cmap='gray')
    plt.title(f'Age: {csv_file_path.loc[idx, "age"]}, Ethnicity: {csv_file_path.loc[idx, "ethnicity"]}, Gender: {csv_file_path.loc[idx, "gender"]}')
    plt.axis('off')
plt.show()
```

```python
# Basic statistics and information about the dataset
print("Basic Dataset Statistics:")
print(csv_file_path.describe())
```

```python
# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(age_counts['Age'], age_counts['Frequency'], color='blue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Frequency Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python

# Age Distribution Histogram
plt.figure(figsize=(10, 6))
plt.hist(csv_file_path['age'], bins=20, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.grid(True)
plt.show()
```

```python
# Ethnicity Distribution Bar Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='ethnicity', data=csv_file_path)
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.title('Ethnicity Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gender Distribution Pie Chart
gender_counts = csv_file_path['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
```
